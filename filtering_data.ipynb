```python
import pandas as pd

# Chargement des fichiers Excel
fichier_1 = 'data/liste ecole dans la base.xlsx'
fichier_2 = 'data/Ecole sourcées avec  masque 3sept.xlsx'

# Lire les fichiers Excel dans des DataFrames
df1 = pd.read_excel(fichier_1)
df2 = pd.read_excel(fichier_2)

# Assurez-vous que les noms de colonnes sont corrects
# Remplacez 'Entreprise / organisation' et 'ecole_name' par les noms de colonnes réels si nécessaire
colonne_fichier_1 = 'Entreprise / organisation'
colonne_fichier_2 = 'ecole_name'

# Obtenir la liste des valeurs uniques dans la colonne 'Entreprise / organisation' de df1
valeurs_df1 = df1[colonne_fichier_1].dropna().unique()

# Filtrer les lignes dans df2 où 'ecole_name' n'existe pas dans 'valeurs_df1'
df2_filtré = df2[~df2[colonne_fichier_2].isin(valeurs_df1)]

# Sauvegarder le DataFrame filtré dans un nouveau fichier Excel
df2_filtré.to_excel('data/Ecole_sourcées_filtré.xlsx', index=False)

print("Le filtrage est terminé et les résultats ont été sauvegardés dans 'Edata/Ecole_sourcées_filtré.xlsx'.")

```
